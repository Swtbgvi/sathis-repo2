```python
# pandas is sql table form datastructure
import pandas as pd
# numpy is a multidimentional array
import numpy as np

df = pd.DataFrame(np.random.randn(10, 4))
print(df)

# break it into pieces
# first 3 rows, then 7+ and 3 to 7
pieces = [df[:3], df[7:], df[3:7],
          pd.DataFrame({'key': ['foo', 'bar'], 'lval': [1, 2]})]
print(pieces)
# concat
print(pd.concat(pieces))

```

```python
# join

#left = pd.DataFrame({'key': ['foo', 'foo'], 'lval': [1, 2]})
#right = pd.DataFrame({'key': ['foo', 'foo'], 'rval': [4, 5]})
left = pd.DataFrame({'key': ['foo', 'bar'], 'lval': [1, 2]})
right = pd.DataFrame({'key': ['foo1', 'bar1'], 'rval': [4, 5]})


print(left)
print(right)

# default inner join
print(pd.merge(left, right, on='key'))
print(pd.merge(left, right, on='key',how='inner'))
print(pd.merge(left, right, on='key',how='left'))
print(pd.merge(left, right, on='key',how='right'))
print(pd.merge(left, right, on='key',how='outer'))
```

```python
# Aggregation
df = pd.DataFrame({'A' : ['foo', 'bar', 'foo', 'bar',
                             'foo', 'bar', 'foo', 'foo'],
                      'B' : ['one', 'one', 'two', 'three',
                             'two', 'two', 'one', 'three'],
                      'C' : np.random.randn(8),
                      'D' : np.random.randn(8)})

print(df)

print(df.groupby('A').sum())
print(df.groupby(['A','B']).sum())
```

```python
# Reshaping
# pandas is sql table form datastructure
import pandas as pd
# numpy is a multidimentional array
import numpy as np

tuples = list(zip(*[['bar', 'bar', 'baz', 'baz',
                        'foo', 'foo', 'qux', 'qux'],
                       ['one', 'two', 'one', 'two',
                        'one', 'two', 'one', 'two']]))
print(tuples)

index = pd.MultiIndex.from_tuples(tuples, names=['first', 'second'])
print(index)
```

```python
df = pd.DataFrame(np.random.randn(8, 2), index=index, columns=['A', 'B'])
print(df)
```

```python
# stack
stacked = df.stack()
print(stacked)
```

```python
stacked.dtypes
```

```python
stacked.unstack()
```

```python
print(stacked.unstack())

print(stacked.unstack(1))

print(stacked.unstack(0))
```

```python
# pivot table

df = pd.DataFrame({'A' : ['one', 'one', 'two', 'three'] * 3,
                      'B' : ['A', 'B', 'C'] * 4,
                      'C' : ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'] * 2,
                      'D' : np.random.randn(12),
                      'E' : np.random.randn(12)})

print(df)

print("Hello")
# index is row field
# column C
# Value is D
print(pd.pivot_table(df, values='D', index=['A', 'B'], columns=['C']))
```
